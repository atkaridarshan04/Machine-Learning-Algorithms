**Epsilon:** radius of the circle within which it searches neighbours

**Core Points:** A point is considered a core point if there are at least minPts (a parameter) points within a distance of epsilon from it, including the point itself.

**Border Points:** A point is considered a border point if it is not a core point but is within the epsilon neighborhood of a core point.

**Noise Points:** Points that are neither core nor border points are considered noise points.

**Density Connectes Points:** Two points that are connected randomly by core points na dthe distance betn the core points < Epsilon

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
X = np.array([[1, 2], [2, 2], [2, 3],[8, 7], [8, 8], [25, 80]])
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
db = DBSCAN(eps=1, min_samples=2)
db.fit(X)
```

```python
db.labels_
```

-1 means noise point, all zeros means they are forming one cluster, all 2 means they are forming another cluster

```python
# Create a concentric circle dataset
X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=4)
```

```python
# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps=0.1, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
dbscan.labels_
```

```python
# Plotting
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Concentric Circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
from sklearn.datasets import make_blobs

# Generate 3 clusters of points
X1, _ = make_blobs(n_samples=150, centers=3, cluster_std=0.5, random_state=0)

# Add some noise points
noise = np.random.uniform(low=-3, high=3, size=(50, 2))
X1 = np.vstack([X1, noise])

# Plot the dataset
plt.figure(figsize=(6, 6))
plt.scatter(X1[:, 0], X1[:, 1], s=10)
plt.show()
```

```python
# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps=0.5, min_samples=8)
clusters = dbscan.fit_predict(X1)
```

```python
dbscan.labels_
```

```python
# Plotting
plt.scatter(X1[:, 0], X1[:, 1], c=clusters, cmap='plasma', marker='o')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
