**Apriori Algorithm uses frequent item sets to generate association rules. It is based on the concept that a subset of a frequent item set must be a frequent item set.**

**Frequent item set is an itemset whose support value is greater than a threshold value.**

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('../DataSets/Online_Retail.csv', encoding='latin1')
df.head()
```

### Preparing Data

```python
df['Description'] = df['Description'].str.strip() # strip is used to remove leading and trailing whitespaces from the values
df.dropna(axis=0, subset=["InvoiceNo"], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')] # to exclude rows where the 'InvoiceNo' column contains the letter 'C'.

df.head()
```

```python
df['Country'].value_counts()
```

```python
df.shape
```

```python
### Seperating one tranction
basket = (df[df['Country'] == 'France']
         .groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
         .unstack().reset_index().fillna(0).set_index('InvoiceNo'))

# df[df['Country'] == 'France']: Filters the original DataFrame df to include only rows where the 'Country' is 'France'.
# .groupby(['InvoiceNo', 'Description'])['Quantity'].sum(): Groups the filtered data by 'InvoiceNo' (invoice number) and 'Description' (product 
# description), then calculates the sum of 'Quantity' for each group. This step essentially aggregates the quantity of each product in each invoice.

# .unstack(): Restructures the grouped data to have 'Description' values as columns, 'InvoiceNo' as index, and the sum of 'Quantity' as values. 
# This transformation pivots the data to a format suitable for market basket analysis, where each row represents an invoice and each column represents a product.

# .reset_index(): Resets the index of the DataFrame, moving 'InvoiceNo' back to a column.
# .fillna(0): Fills any missing values with 0. This step ensures that if a product didn't appear in an invoice, its quantity is considered as 0.
# .set_index('InvoiceNo'): Sets the 'InvoiceNo' column as the index of the DataFrame, making it the row identifier for each invoice.

basket.head()
```

```python
### Converting all positive values to 1 and everything else to 0
def encode_units(x):
    if x >= 1:
        return 1
    else:
        return 0

basket_sets = basket.map(encode_units).astype(bool)
basket_sets.drop('POSTAGE', inplace=True, axis=1) 

basket_sets.head()
```

## Training Apriori Algorithm

### Generating Itemsets

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.head()
```

### Making Rules based on conditions

```python
rules[ (rules['lift'] >= 6) & rules['confidence'] >= 0.8]
```
